# Query 1
What are the percentages of each rating digit for the product with id: 21?

```
reviews = spark.table('reviews')
display(reviews)
```

```
# Query for percent of all ratings for product ID 21
query = '''
SELECT r.rating,
  COUNT(*) AS total_ratings,
  ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (), 2) AS percentage
FROM
  products p
  INNER JOIN reviews r ON p.product_id = r.review_product_id
WHERE
  p.product_id = 'pid_21'
GROUP BY
  r.rating;

'''
id_21_ratings = spark.sql(query)
id_21_ratings.show()
```
