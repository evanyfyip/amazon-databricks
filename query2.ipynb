# Query 2
Which pairs of products have consistently stayed within each other’s “Customers who bought this item also bought lists through March, May and June of 2003?

```
graphs = spark.table('graphs')
display(graphs)
```

```
query = '''
SELECT DISTINCT collection_date
FROM graphs
ORDER BY collection_date::date;
'''
display(spark.sql(query))
```

```
query = '''
SELECT DISTINCT
    t1.from_id,
    t1.to_id
FROM graphs t1
INNER JOIN graphs t2 ON
    t1.from_id = t2.to_id AND
    t1.to_id = t2.from_id AND
    t1.collection_date = t2.collection_date
GROUP BY t1.from_id, t1.to_id
HAVING COUNT(t1.collection_date) = 4
'''
query_results = spark.sql(query)
display(query_results.count())
```

```
graphs_0302 = graphs.filter("collection_date == '2003-03-02'")
graphs_0312 = graphs.filter("collection_date == '2003-03-12'")
graphs_0505 = graphs.filter("collection_date == '2003-05-05'")
graphs_0601 = graphs.filter("collection_date == '2003-06-01'")

vertices_0302 = graphs_0302.select('from_id').withColumnRenamed('from_id', 'id') \
    .union(graphs_0302.select('to_id').withColumnRenamed('to_id', 'id')).distinct()
vertices_0312 = graphs_0312.select('from_id').withColumnRenamed('from_id', 'id') \
    .union(graphs_0312.select('to_id').withColumnRenamed('to_id', 'id')).distinct()
vertices_0505 = graphs_0505.select('from_id').withColumnRenamed('from_id', 'id') \
    .union(graphs_0505.select('to_id').withColumnRenamed('to_id', 'id')).distinct()
vertices_0601 = graphs_0601.select('from_id').withColumnRenamed('from_id', 'id') \
    .union(graphs_0601.select('to_id').withColumnRenamed('to_id', 'id')).distinct()

edges_0302 = graphs_0302.select('from_id', 'to_id').withColumnRenamed('from_id', 'src').withColumnRenamed('to_id', 'dst')
edges_0312 = graphs_0312.select('from_id', 'to_id').withColumnRenamed('from_id', 'src').withColumnRenamed('to_id', 'dst')
edges_0505 = graphs_0505.select('from_id', 'to_id').withColumnRenamed('from_id', 'src').withColumnRenamed('to_id', 'dst')
edges_0601 = graphs_0601.select('from_id', 'to_id').withColumnRenamed('from_id', 'src').withColumnRenamed('to_id', 'dst')

gf_0302 = GraphFrame(vertices_0302, edges_0302)
gf_0312 = GraphFrame(vertices_0312, edges_0312)
gf_0505 = GraphFrame(vertices_0505, edges_0505)
gf_0601 = GraphFrame(vertices_0601, edges_0601)

pairs_0302 = gf_0302.find('(a)-[e]->(b); (b)-[e2]->(a)')
pairs_0312 = gf_0312.find('(a)-[e]->(b); (b)-[e2]->(a)')
pairs_0505 = gf_0505.find('(a)-[e]->(b); (b)-[e2]->(a)')
pairs_0601 = gf_0601.find('(a)-[e]->(b); (b)-[e2]->(a)')
common_pairs = pairs_0302.intersect(pairs_0312).intersect(pairs_0505).intersect(pairs_0601)
display(common_pairs.count())
```
